### Activity 2: Explore contour() and contourf()

In the following cells, you will explore the `contour()` and `contourf()` plotting functions with new arrays

```python
## CELL 2A 
## INSTRUCTIONS: Run this cell

# Create two new fields based on the original x and y data from above
x = np.arange(-5.0, 5.0, 0.1)
y = np.arange(-5.0, 5.0, 0.1)
X, Y = np.meshgrid(x, y)

cosZ = np.cos(X) + np.cos(Y)
sinZ = np.sin(X) + np.sin(Y)
```

```python
## CELL 2B
## INSTRUCTIONS: Predict what the output of this cell will look like, then run this cell

plt.contourf(cosZ)
plt.contour(sinZ)
```

Matplotlib by default will continue adding plots atop the same axes, more on this later. We see that cosZ and sinZ are offset by 90 degrees in both x and y from each other, as expected. We are able to see this because we layered a `contour` plot atop a `contourf` plot. 

Contours can even be plotted over an imshow plot.

#### Colors and colormaps

When layering grids on top of each other, we often find the need to use different colormaps to more easily distinguish the different layers. We can change the colormap by adding the `cmap` argument to the imshow, contour, and contourf functions. 


Examples:
>
    plt.imshow(Z, cmap="coolwarm")
    plt.contour(cosZ, cmap="gray")

Click this link to see all of the valid matplotlib colormap options:
